# Simple Model
Import packages, warning message can be ignored

```python
import pandas as pd
import matplotlib.pyplot as plt
import hddm
# Load data from csv file into a NumPy structured array
data = hddm.load_csv('C:\Users\Edwin\Documents\UCR_ATdata_Discrim_N17.csv')
```

Load data and create a model object before MCMC sampling to fit the model

```python
# Create a HDDM model multi object
model = hddm.HDDM(data)

# find a good starting point
model.find_starting_values()

# Create model and start MCMC sampling
model.sample(2000, burn=20)
```

```python
# Print fitted parameters and other model statistics
model.print_stats()
```

```python
# Plot posterior distributions and theoretical RT distributions
model.plot_posteriors(['a', 't', 'v', 'a_std'])
```

Plots for a and v look acceptable, a_std is a bit suspicious, t is just bad <br>
Next we look the Gelman-Rubin statistic to check the intra-chain variance of different instances of the model

```python
models = []
for i in range(5):
    m = hddm.HDDM(data)
    m.find_starting_values()
    m.sample(5000, burn=20)
    models.append(m)
hddm.analyze.gelman_rubin(models)
```

```python
model.plot_posterior_predictive(figsize=(14,10))
```

The model doesn't fit the data all that well, there is probably a correlation between types of stimuli 

```python
model_stim = hddm.HDDM(data, depends_on={'v':'stimName'})
model_stim.find_starting_values()
model_stim.sample(5000, burn=1000)
```

```python
model_stim.plot_posteriors(['a', 't', 'v', 'a_std'])
```

```python
v_A, v_T, v_AT = model_stim.nodes_db.node[['v(A)', 'v(T)', 'v(AT)']]
hddm.analyze.plot_posterior_nodes([v_A, v_T, v_AT])
```

# Within Subject Model

```python
from patsy import dmatrix
dmatrix("stimName", data.head(10))
```

```python
model_withinsubj = hddm.HDDMRegressor(data, "v ~ C(stimName, Treatment('AT'))")
model_withinsubj.sample(5000, burn=200)
```

```python
v_A, v_T, v_AT = model_withinsubj.nodes_db.ix[['v_Intercept', "v_C(stimName, Treatment('AT'))[T.A]", "v_C(stimName, Treatment('AT'))[T.T]"], 'node']
hddm.analyze.plot_posterior_nodes([v_A, v_T, v_AT])
plt.xlabel('drift-rate')
plt.ylabel('Posterior probability')
```

```python
model_withinsubj.plot_posteriors()
```

To improve the diagnostic plots, try sampling more (10000 instead of 5000)
